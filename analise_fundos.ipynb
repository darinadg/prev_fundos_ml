```python
ipca["data"] = ipca["data"].astype(str)

ipca["ano"] = ipca["data"].str[:4]
ipca["mes"] = ipca["data"].str[5:]

ipca["data"] = ipca["ano"] + "-" + ipca["mes"] + "-01"

ipca["data"] = pd.to_datetime(ipca["data"])

ipca = ipca[["data", "ipca"]]

ipca.head()
```

```python
selic = pd.read_excel("ipeadata[10-06-2015-03-39]TAXAJUROS-SELIC.xls", skiprows=10)

selic.columns = ["data", "selic"]

# converter direto para data
selic["data"] = pd.to_datetime(selic["data"], dayfirst=True)

selic.head()
```

```python
pib = pd.read_excel("ipeadata[10-06-2015-03-55]PIB-90-4.15.xls", skiprows=10)

pib.columns = ["data", "pib"]

# transformar em string
pib["data"] = pib["data"].astype(str)

# separar ano e período
pib["ano"] = pib["data"].str[:4]
pib["periodo"] = pib["data"].str[5:]

# transformar período em mês (aproximação)
# 1 -> jan, 2 -> abr, 3 -> jul, 4 -> out
mapa = {
    "01": "01",
    "02": "04",
    "03": "07",
    "04": "10"
}

pib["mes"] = pib["periodo"].map(mapa)

# criar data
pib["data"] = pib["ano"] + "-" + pib["mes"] + "-01"
pib["data"] = pd.to_datetime(pib["data"])

# manter colunas finais
pib = pib[["data", "pib"]]

pib = pib.dropna()
pib.head()
```

```python
poupanca = pd.read_excel("ipeadata[10-06-2015-03-Poup.Nac.Bruta 91-14 trimestral.xls", skiprows=10)

poupanca.columns = ["data", "poupanca"]

# transformar em string
poupanca["data"] = poupanca["data"].astype(str)

# separar ano e trimestre
poupanca["ano"] = poupanca["data"].str[:4]
poupanca["trimestre"] = poupanca["data"].str[-2:]  # pega T1, T2...

# mapa correto
mapa = {
    "T1": "01",
    "T2": "04",
    "T3": "07",
    "T4": "10"
}

poupanca["mes"] = poupanca["trimestre"].map(mapa)

# criar data
poupanca["data"] = poupanca["ano"] + "-" + poupanca["mes"] + "-01"
poupanca["data"] = pd.to_datetime(poupanca["data"])

# manter colunas finais
poupanca = poupanca[["data", "poupanca"]]

# remover possíveis erros
poupanca = poupanca.dropna()

poupanca.head()
```

```python
# SELIC mensal
selic["data"] = selic["data"].dt.to_period("M").dt.to_timestamp()

selic_mensal = selic.groupby("data")["selic"].mean().reset_index()

# PIB mensal
pib_mensal = pib.set_index("data").resample("M").ffill().reset_index()

# POUPANÇA mensal
poupanca_mensal = poupanca.set_index("data").resample("M").ffill().reset_index()

# IPCA já mensal
ipca_mensal = ipca.copy()

# juntar tudo
df_final = ipca_mensal.merge(selic_mensal, on="data") \
                     .merge(pib_mensal, on="data") \
                     .merge(poupanca_mensal, on="data")

df_final.head()
```
